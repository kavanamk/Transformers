get_stats takes in input the list of tokens

*   then counts the frequency of each token pair
*   returns the count

```python
def get_stats(ids):
    counts = {}
    for i in range(len(ids)-1):
      if (ids[i],ids[i+1]) in counts:
        counts[(ids[i],ids[i+1])] += 1
      else:
        counts[(ids[i],ids[i+1])] = 1

    return counts

stats = get_stats(tokens)
print(stats)
# print(sorted(((v,k) for k,v in stats.items()), reverse=True))
```

convert the dictionary into a maz heap

```python
import heapq
def get_heap(stats):
  heap=[]
  for key,val in stats.items():
    heap.append((-val,key))
  heapq.heapify(heap)
  return heap
heap= get_heap(stats)
print(heap)
```

get the top of this heap

```python
import heapq
def get_top_pair(heap_obj):
  return heapq.heappop(heap_obj)
_,top_pair=get_top_pair(heap)
print(top_pair)
```

create a pipeline for -> dict-> heap-> top value-> merge

```python
class Encoder:
  def __init__(self, text):
    self.text = text
    # Convert text to initial tokens once during initialization
    raw_bytes = text.encode("utf-8")
    self.tokens = list(map(int, raw_bytes))
    self.vocab = {}

  def get_stats(self,ids):
    counts = {}
    for i in range(len(ids)-1):
      if (ids[i],ids[i+1]) in counts:
        counts[(ids[i],ids[i+1])] += 1
      else:
        counts[(ids[i],ids[i+1])] = 1

    return counts

  def get_heap(self,stats):
    heap=[]
    for key,val in stats.items():
      heap.append((-val,key))
      heapq.heapify(heap)
      return heap

  def get_top_pair(self, heap_obj):
    return heapq.heappop(heap_obj)

  def merge(self, ids, pair, idx):
    new_ids = []
    i = 0
    while i < len(ids):
        if i + 1 < len(ids) and (ids[i], ids[i+1]) == pair:
            new_ids.append(idx)
            i += 2 # Skip the second element of the pair as it's merged
        else:
            new_ids.append(ids[i])
            i += 1
    return new_ids

  def encode(self, iter):
    # Work on a copy of the initial tokens, or an evolving internal state
    current_ids = list(self.tokens) # Start with the initial tokens
    next_idx = 9999 # Using 9999 as per your previous choice
    for _ in range(iter):
      stats = get_stats(current_ids) # Use current_ids
      # Generate a new heap for the current iteration based on the current stats
      current_heap = get_heap(stats)

      if not current_heap: # If no pairs are found, break the loop
          break

      # Get the top pair from the current_heap.
      # get_top_pair modifies the heap by popping the smallest element.
      # The tuple is (-count, pair), so we take [1] for the pair.
      _, top_pair = get_top_pair(current_heap)

      current_ids = merge(current_ids, top_pair, idx=next_idx) # Update current_ids
      self.vocab[next_idx] = top_pair
      next_idx += 1
    return current_ids # Return the final encoded IDs

  def getVocabulary(self):
    return self.vocab
```

```python
# Instantiate the Encoder class
encoder = Encoder(text)

# Encode the text for 10 iterations
encoded_tokens = encoder.encode(10)

# Print the encoded tokens
print("Encoded Tokens:", encoded_tokens)

# Print the learned vocabulary
print("Learned Vocabulary:", encoder.getVocabulary())
```

Decoder

```python
class Decoder:
  def __init__(self, vocab):
    self.vocab = vocab
    # Create a reverse mapping for merged tokens to their constituent parts
    # This helps in the iterative decoding process
    self.merges = {idx: pair for idx, pair in vocab.items()}

  def decode(self, ids):
    """
    Decodes a list of token IDs back into the original text.

    Args:
      ids: A list of integers representing the encoded tokens.

    Returns:
      A string representing the decoded text.
    """
    current_tokens = list(ids)

    # Continue expanding merged tokens until no more merged tokens (IDs > 255) are present.
    # The `any` check will be true if there's at least one token ID greater than 255,
    # indicating a merged token that needs to be expanded.
    while any(token_id > 255 for token_id in current_tokens):
        new_tokens_pass = []
        for token_id in current_tokens:
            if token_id in self.merges:
                # If the token_id is a merged token, replace it with its constituent parts
                part1, part2 = self.merges[token_id]
                new_tokens_pass.append(part1)
                new_tokens_pass.append(part2)
            else:
                # If it's a base byte token (0-255), keep it as is
                new_tokens_pass.append(token_id)
        current_tokens = new_tokens_pass

    # After the loop, current_tokens should only contain byte values (0-255).
    # Convert the list of integers to a bytes object and then decode it to a UTF-8 string.
    byte_array = bytes(current_tokens)
    return byte_array.decode("utf-8", errors="replace")
```

```python
# Instantiate the Decoder class with the learned vocabulary
decoder = Decoder(encoder.getVocabulary())

# Decode the encoded tokens
decoded_text = decoder.decode(encoded_tokens)

# Print the decoded text
print("Decoded Text:")
print(decoded_text)

# Verify if the decoded text matches the original text
print("\nDoes decoded text match original text?", decoded_text == text)
```
